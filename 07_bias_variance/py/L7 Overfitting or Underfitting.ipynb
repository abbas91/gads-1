```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
import warnings
import seaborn as sns; 
sns.set(style="ticks", color_codes=True)
warnings.simplefilter(action = "ignore", category = FutureWarning)

%matplotlib inline

df = pd.read_csv('../data/housing.data.csv')
df.corr()
```

```python
from sklearn.metrics import mean_squared_error

np.random.seed(0)

df = pd.read_csv('../data/housing.data.csv')

# Shuffle
df = df.iloc[np.random.permutation(len(df))]
df.reset_index(drop=True)

# train set: the first 100 rows
df_train = df[:100]
# df_train.sort_values(by = 'NOX', inplace = True)

# test set: the first 100 rows
df_test = df[100:]
print(df_test.shape)

# set aside the test set
# format the train set for use with scikit

X_train = df_train.NOX.reshape(-1,1)
y_train = df_train.DIS

X_test = df_test.NOX.reshape(-1,1)
y_test = df_test.DIS
```

```python
# 1st regression model simple linear regression model

clf_1 = linear_model.LinearRegression()
clf_1.fit(X_train,y_train)

# error on the training set
y_hat_train_1 = clf_1.predict(X_train) 

e_train_1 = mean_squared_error(y_hat_train_1, y_train)
print("MSE for training on deg 1 classifier : %s "% e_train_1)

# error on the test set
y_hat_test_1 = clf_1.predict(X_test) 
e_test_1 = mean_squared_error(y_hat_test_1, y_test)
print("MSE for test on deg 1 classifier : %s "% e_test_1)
```

```python
# 2nd model: polynomial degree 4

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(4)
X_train_4 = poly.fit_transform(X_train)
X_test_4 = poly.fit_transform(X_test)

print(X_train_4.shape)

clf_4 = linear_model.LinearRegression()
clf_4.fit(X_train_4,y_train)

# error on the training set
y_hat_train_4 = clf_4.predict(X_train_4) 

e_train_4 = mean_squared_error(y_hat_train_4, y_train)
print("MSE for training on deg 1 classifier : %s "% e_train_4)

# error on the test set
y_hat_test_4 = clf_4.predict(X_test_4) 
e_test_4 = mean_squared_error(y_hat_test_4, y_test)
print("MSE for test on deg 4 classifier : %s "% e_test_4)
```

```python
# 3nd model: polynomial degree 16
poly = PolynomialFeatures(16)
X_train_16 = poly.fit_transform(X_train)
X_test_16 = poly.fit_transform(X_test)

print(X_train_16.shape)

clf_16 = linear_model.LinearRegression()
clf_16.fit(X_train_16,y_train)

# error on the training set
y_hat_train_16 = clf_16.predict(X_train_16) 

e_train_16 = mean_squared_error(y_hat_train_16, y_train)
print("MSE for training on deg 1 classifier : %s "% e_train_16)

# error on the test set
y_hat_test_16 = clf_16.predict(X_test_16) 
e_test_16 = mean_squared_error(y_hat_test_16, y_test)
print("MSE for test on deg 16 classifier : %s "% e_test_16)
```
