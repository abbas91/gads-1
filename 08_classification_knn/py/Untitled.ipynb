```python
from sklearn import datasets, neighbors, metrics
import pandas as pd
import seaborn as sns

%matplotlib inline

iris = datasets.load_iris()

irisdf = pd.DataFrame(iris.data, columns=iris.feature_names)
irisdf['target'] = iris.target

cmap = {'0': 'r', '1': 'y', '2': 'b' }
irisdf['ctarget'] = irisdf.target.apply(lambda x: cmap[str(x)])

irisdf.plot('petal length (cm)', 'petal width (cm)', kind='scatter', c=irisdf.ctarget)

```

```python
threshold = 2.5

def classifier(x,  threshold, true_class, false_class):
    if x > threshold:
        return true_class
    else:
        return false_class
    


y_hat = irisdf['petal length (cm)'].apply(lambda x : classifier(x) )

metrics.accuracy_score(y_hat, iris.target)
```
