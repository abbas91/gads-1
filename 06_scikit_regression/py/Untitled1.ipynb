```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)

%matplotlib inline
```

```python
df = pd.read_csv('../data/polynomial.csv', index_col=None)
X = df.x.reshape(-1, 1)
y = df.y

# first case linear

clf = linear_model.LinearRegression()
clf.fit(X,y)
y_hat1 = clf.predict(X)
print("MSE (deg=1) %s"% mean_squared_error(y, y_hat1) )

# second case Adding a quadratic term x + x^2
poly = PolynomialFeatures(3)
X2 = poly.fit_transform(X)

clf = linear_model.LinearRegression()
clf.fit(X2,y)
y_hat2 = clf.predict(X2)
print("MSE (deg=2) %s"% mean_squared_error(y, y_hat2) )


# third case high order polynomial
poly = PolynomialFeatures(16)
X16 = poly.fit_transform(X)

clf = linear_model.LinearRegression()
clf.fit(X16,y)
y_hat16 = clf.predict(X16)
print("MSE (deg=16) %s"% mean_squared_error(y, y_hat16) )

# third case high order polynomial
poly = PolynomialFeatures(20)
X20 = poly.fit_transform(X)

clf = linear_model.LinearRegression()
clf.fit(X20,y)
y_hat20 = clf.predict(X20)
print("MSE (deg=20) %s"% mean_squared_error(y, y_hat20) )

```

```python
fig,[ax1,ax2, ax3] = plt.subplots(1,3,figsize=(18,6))

ax1.plot(y,'.', c='b', label='truth')
ax1.plot(y_hat1, c='r', label='prediction')
ax1.set_title("Simple regression")
ax1.annotate("RMSE %0.2f"% np.sqrt(mean_squared_error(y, y_hat1)), xy = (5,150) )
ax1.legend(loc='best')

ax2.plot( y,'.', c='b', label='truth')
ax2.plot(y_hat16, c='r', label='prediction')
ax2.set_title("Polynomial (deg = 16) regression")
ax2.annotate("RMSE %0.2f"% np.sqrt(mean_squared_error(y, y_hat16)), xy = (5,150) )
ax2.legend(loc='best')

ax3.plot( y,'.', c='b', label='truth')
ax3.plot(y_hat20, c='r', label='prediction')
ax3.set_title("Polynomial (deg = 20) regression")
ax3.annotate("RMSE %0.2f"% np.sqrt(mean_squared_error(y, y_hat20)), xy = (5,150) )
ax3.legend(loc='best')
```
