```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)

from sklearn import linear_model
```

```python
np.random.seed(88)
n = 100
idx = np.linspace(0,n-1,n)

X = pd.DataFrame( {'idx': np.linspace(0,n-1,n), 'vals': np.linspace(0,n-1,n) + np.random.random(n) } )

y = np.linspace(3,n + 2,n) + np.random.normal(0,n/10,n)

plt.plot(X, y,'+')

```

```python
clf = linear_model.LinearRegression(fit_intercept= True)
clf.fit(X,y)
clf.coef_

y_hat_scikit = clf.predict(X)

y_hat_alg = np.array(X).dot(clf.coef_)

plt.plot(y_hat_alg  - y_hat_scikit)
clf.intercept_
```

```python
fig,[ax1,ax2] = plt.subplots(1,2,figsize=(12,6))

# True outcomes in blue
ax1.plot(X.vals, y,'.', c='b', label='truth')
# # Predicted values in red
ax1.plot(X.vals, y_hat_scikit,'+', c='r', label='prediction')

ax1.set_title("Prediction vs Truth")
ax1.legend(loc='best')

ax2.plot(X.vals,y-y_hat_scikit,'.', c='b', label='Residuals')
ax2.set_title("Residuals")
ax2.legend(loc='best')



```
