```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)


df = pd.read_csv('../data/ozone.csv')
df.shape
```

```python
print(df.head())
print(df.shape)
df = df.dropna()
print(df.shape)
```

```python
fig, ax = plt.subplots(2,2, figsize=(9,9))

plt.subplot(221)
ax = plt.gca()
ax.set_axis_bgcolor('white')
df.Ozone.plot(kind='hist', label="hist", legend=True, bins = 30, secondary_y=True)

df.Ozone.plot(kind='density', label="density", legend=True)

plt.title('Ozone')
plt.grid()
# 
plt.subplot(222)
ax = plt.gca()
ax.set_axis_bgcolor('white')
df.Temp.plot(kind='hist', label="hist", legend=True, bins = 30, secondary_y=True)

df.Temp.plot(kind='density', label="density", legend=True)

plt.title('Temp')
plt.grid()

```
