```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)


df = pd.read_csv('../data/ozone.csv')
df.shape
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
# Using plt.hist, draw 4 subplots, each with histogram and density
# labeled, titled, nice axes, ...
```

```python
fig, ax = plt.subplots(2,2, figsize=(12,12))


# 1st subplot
plt.subplot(221)
ax = plt.gca()
ax.set_axis_bgcolor('white')
df.Ozone.plot(kind='hist', secondary_y=True, label="hist", legend=True, bins = 30)
df.Ozone.plot(kind='density', label="density", legend=True)
plt.title('Ozone')
plt.grid()

# 2nd subplot

plt.subplot(222)
ax = plt.gca()
ax.set_axis_bgcolor('white')
df.Wind.plot(kind='hist', secondary_y=True, label="hist", legend=True, bins = 30)
df.Wind.plot(kind='density', label="density", legend=True)
plt.title('Wind')
plt.grid()

# 3rd subplot

plt.subplot(223)
ax = plt.gca()
ax.set_axis_bgcolor('white')
df['Solar.R'].plot(kind='hist', secondary_y=True, label="hist", legend=True, bins = 30)
df['Solar.R'].plot(kind='density', label="density", legend=True)
plt.title('Solar.R')
plt.grid()

# 4th subplot

plt.subplot(224)
ax = plt.gca()
ax.set_axis_bgcolor('white')
df.Temp.plot(kind='hist', secondary_y=True, label="hist", legend=True, bins = 30)
df.Temp.plot(kind='density', label="density", legend=True)
plt.title('Temp')
plt.grid()


```

```python
df[['Ozone','Solar.R','Wind','Temp']].corr()
```

```python
# Scatter plot with Temp vs Ozone

fig = plt.figure()

plt.scatter(df.Ozone, df.Wind, s=150,  c=df.Temp, cmap=plt.cm.coolwarm)

# plt.grid()
ax = plt.gca()
ax.set_axis_bgcolor('white')
plt.xlabel('Wind')
plt.ylabel('Ozone')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')

plt.show()
```
