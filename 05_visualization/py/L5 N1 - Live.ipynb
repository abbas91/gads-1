```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.linspace(-np.pi, np.pi, 256,endpoint=True)

C = np.cos(X)

S = np.sin(X)
```

```python
plt.plot(C)
plt.plot(S)
```

```python
np.random.seed(888)
x,y = np.random.randn(2, 100)


# Create the figure
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_subplot(211)

ax2 = fig.add_subplot(212)


# Work on the first subplot
ax1.plot(x)
ax1.set_title('This is X')

# # Work on the second subplot
ax2.plot(y)
ax2.set_title('This is Y')

# # set titles for each subplots
```

```python
fig = plt.figure()

plt.subplot(211)

plt.plot(x)

# All operations are on the first  plot
plt.title('this is X')


plt.subplot(212)
plt.plot(y)
plt.title('this is Y')

plt.subplot(211)

plt.plot(x + y)
```

```python
fig = plt.figure(figsize=(9,6))
# First subplot
ax1 = fig.add_subplot(211)

# get the position of 1st subplot
pos1 = ax1.get_position() 

print(pos1)

# New coordinates: simple translatation
pos2 = [pos1.x0 + 1, pos1.y0 + 0.2,  pos1.width , pos1.height] 
# # New coordinates: translatation and contraction
pos2 = [pos1.x0 - 0.05 , pos1.y0 + 0.15 ,  pos1.width / 3 , pos1.height /2] 
print(pos2)


# # 2nd subplot
ax2 = fig.add_subplot(221)

# set 2nd subplot position
ax2.set_position(pos2) 
ax2.grid()


```

```python
fig, ax = plt.subplots(figsize=(6, 4))

plt.plot(X,C,  color="blue", linewidth=1.0, linestyle="-", label="Cosine")
plt.plot(X,S,  color="red", linewidth=3.0, linestyle="dashed", label = "Sine")

plt.ylim(-1.2, 1.2)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']  )

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))


ax.legend(loc='best', frameon=False)

# annotations
t = 2*np.pi/3
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), 
             xycoords='data',
             xytext=(+10, +30), 
             textcoords='offset points', 
             fontsize=20,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
            )

plt.title('Trigonometry')

fig.savefig('lesson_6_Trigonometry_101.png')
```
